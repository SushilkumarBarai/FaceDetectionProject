```python
import cv2
print('opencv %s' % cv2.__version__)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2.imshow('Your Live Fram', frame)
    #print(ret)
    #print(frame)
    if cv2.waitKey(1)==13:
        break

cap.release()
cv2.destroyAllWindows()
```
